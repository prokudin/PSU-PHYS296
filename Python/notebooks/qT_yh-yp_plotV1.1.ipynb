```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
```

```python
compass_data = pd.read_excel("../database/sidis/expdata/5001.xlsx")
```

```python
compass_data_pip = compass_data[compass_data["hadron"]=="pi+"]
```

```python
compass_data_pip['qT'] = compass_data_pip['pT']/compass_data_pip['z']
compass_data_pip['yh-yp'] = compass_data_pip['yh']-compass_data_pip['yp']
```

```python
compass_data_pip_filtered = compass_data_pip.drop([
      'Ebeam','obs','value','stat_u','sys_u','target','hadron','col'], axis=1)
Q = np.sqrt(compass_data_pip_filtered['Q2'])
dat = compass_data_pip_filtered[compass_data_pip_filtered['qT'] < 0.5*Q]
```

```python
plt.figure(1)
sns.relplot(x="yh-yp", y="qT", hue="x",palette= sns.hls_palette(22, l=.4, s=.8), data=dat,legend='full',aspect = 1.5)
plt.xlabel('yh-yp',fontsize = 14,fontweight = 500)
plt.ylabel('qT',fontsize = 14,fontweight = 500,rotation = True)
plt.title('qT vs. yh-yp',fontsize = 14,fontweight = 500)
```

```python
plt.figure(2)
sns.relplot(x="yh-yp", y="qT", hue="x",palette= sns.color_palette("inferno", n_colors=22, desat=.99), data=dat,legend='full',aspect = 1.5)
plt.xlabel('yh-yp',fontsize = 14,fontweight = 500)
plt.ylabel('qT',fontsize = 14,fontweight = 500,rotation = True)
plt.title('qT vs. yh-yp',fontsize = 14,fontweight = 500)
```

```python
# Define a palette with n # colors interpolated to 22
# Note you can define the palette with more than 3 then interpolate to 22
cmap = sns.blend_palette(["#c80815",'#2E37FE', "#00a86b"], 22,as_cmap = False) 
sns.set_palette(cmap, n_colors=22)
# Plots with new palette
plt.figure(3)
sns.relplot(x="yh-yp", y="qT", hue="x",palette= cmap, data=dat,legend='full',aspect = 1.5)
plt.xlabel('yh-yp',fontsize = 14,fontweight = 500)
plt.ylabel('qT',fontsize = 14,fontweight = 500,rotation = True)
plt.title('qT vs. yh-yp',fontsize = 14,fontweight = 500)
```

```python
sns.palplot(cmap)
# Line of code below allows the saturation and number of colors to change if desired
sns.palplot(sns.color_palette(cmap, n_colors=22, desat=.75))
```

```python
plt.figure(4)
sns.relplot(x="yh-yp", y="qT", hue="x",palette= sns.color_palette("GnBu_d", 22), data=dat,legend='full',aspect = 1.5)
plt.xlabel('yh-yp',fontsize = 14,fontweight = 500)
plt.ylabel('qT',fontsize = 14,fontweight = 500,rotation = True)
plt.title('qT vs. yh-yp',fontsize = 14,fontweight = 500)
```
